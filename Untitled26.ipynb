```julia
f = open("VitualBitmap01.txt","r")
n = countlines(f)
seekstart(f)
lista1 = []
listb1 = []
for i = 1:n
    a,b = split(readline(f),",")
    a = parse(Int64,a)
    b = parse(Float64,b)
    append!(lista1,a)
    append!(listb1,b)
end
close(f)
```
